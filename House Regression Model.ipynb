```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
house = pd.read_csv('house.csv')
house
```

```python
house.dropna(inplace=True)
```

```python
house
```

```python
house.drop(columns=['sqft_lot15','lat','sqft_lot','long','sqft_living15','waterfront','view'], axis=1, inplace=True)
```

```python
house
```

```python
house.info()
```

```python
house['bathrooms'] = house['bathrooms'].astype(int)
```

```python
house.info()
```

```python
house['floors'] = house['floors'].astype(int)
```

```python
house.info()
```

```python
house.reset_index(drop=True)
```

```python
X = house.drop(['price', 'yr_built', 'yr_renovated', 'zipcode', 'sqft_above', 'sqft_basement'], axis=1)
y = house['price']
y = y.values.reshape(-1,1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=234)
```

```python
from sklearn.metrics import r2_score
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
r2_score(y_test, y_pred)
```

```python
score = []
for i in range(1000):
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=i)
    lr = LinearRegression()
    lr.fit(X_train, y_train)
    y_pred = lr.predict(X_test)
    score.append(r2_score(y_test, y_pred))
```
